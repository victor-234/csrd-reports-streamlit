```python
import pandas as pd
import requests
```

```python
headers = {"Content-Type": "application/json"}
data = {
    "query": "What are the transition plans of these companies according to ESRS E1?",
    "year": 2024,
    "company": "Airbus"
}

response = requests.post(
    "https://sunhat-api.onrender.com/sustainability-reports/query", 
    json=data, 
    headers=headers
    )

sorted(response.json(), key=lambda x: x["score"], reverse=True)
```

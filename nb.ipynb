```python
import pandas as pd
import requests
```

```python
headers = {"Content-Type": "application/json"}
data = {
    "query": "What are the transition plans of these companies according to ESRS E1?",
    "year": 2024,
    "name": "HSBC Continental Europe"
    # "sector": "Commercial Banks"
}

response = requests.post(
    "https://sunhat-api.onrender.com/sustainability-reports/query", 
    json=data, 
    headers=headers
    )

(
    pd.DataFrame(response.json())
    .sort_values("reportId")
    .assign(
        companyName = lambda x: [requests.get(f"https://sunhat-api.onrender.com/sustainability-reports/reports/{reportId}").json()["company"]["name"] for reportId in x.reportId]
        )
)
```
